# 문제 3

다음 데이터에 사용하여 각 문항에 따라 답하시오.

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

### 문제 [3-1] (10점)

titanic에서 pclass, age, sibsp, fare, class 열만 뽑아 df에 저장하시오.

```python
df = titanic[['pclass', 'age', 'sibsp', 'fare', 'class']]
df = df.loc[df['sibsp'] == 0]
# df = titanic.loc[titanic.sibsp==0, [열 목록]] 이렇게 해도 ok.
df.reset_index(drop= True, inplace=True)
# drop : 기존 index로 사용하던 열을 삭제할지 여부
df.head()
```

### 문제 [3-2] (5점)

df의 열 별 결측치 개수를 구하시오.

```python
# 열 별 결측치 구하기
df.isnull().sum()
```

### 문제 [3-3] (10점)

df에 결측치가 있다면, 해당 열의 평균값으로 채우시오.

```python
# 주의! 
df.fillna(value=df.mean(), inplace=True)
df.head()
```

### 문제 [3-4] (5점)

survived.csv를 불러와 survived에 저장하시오.

```python
survived = pd.read_csv('survived.csv')
survived
```

### 문제 [3-5] (10점)

기존에 생성한 df와 survived 데이터프레임을 병합해 df2에 저장하시오.

```python
df2 = pd.concat([df, survived], axis=1)
df2.head()
```

### 문제 [3-6] (20점)

df2 데이터에서 각 class 별 생존율을 구하시오.

```python
# 이렇게 해도 ok.
# df2.groupby('class').sum()['survived'] / df2.groupby('class').count()['survived']

# 그룹 연산에서 사용자 정의 함수 적용
df2.groupby('class')['survived'].agg(lambda x: x.sum() / len(x))
```
